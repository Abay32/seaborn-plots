# Seaborn tutorial

```python
# Import packeges 
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
%reload_ext autoreload
%autoreload  2

# Import data from sns library
df = sns.load_dataset('car_crashes')
df.head()
```

# Distribution plot


```python
# Univariate distribution 
sns.distplot(df['not_distracted'], kde=True, bins=25)
#kde comulative distribution estimation
#bins is used as a pocket
```

# Joint plot 


```python
sns.jointplot(x='speeding', y='alcohol', data=df, kind='reg')
```

# KDE plot


```python
sns.kdeplot(df['alcohol']) # Comulativedistribution density 
```

# Pair plot 


```python
sns.pairplot(df)

tips = sns.load_dataset('tips')
# show the attributes relationship and show the genders distribution
sns.pairplot(tips, hue='sex')
```

# Rug plot

```python
sns.rugplot(tips['tip'])
```

# Styling the plot

```python
sns.set_style('dark')


sns.set_context('talk', font_scale=1.4) #used to control the font size and type of the figure being used, eg 'paper' for paper, 'talk' for presentations, 'poster' for poster presentation
plt.figure(figsize= (8, 4)) # Figure size determination

sns.jointplot(x='speeding', y='alcohol', data= df, kind='reg')
```

# Categorical plots 

## Bar plot

```python
sns.barplot(x='sex', y = 'total_bill', data = tips, estimator= np.median)
# estimator is aggregate function such as: 'mean', 'median', 'cov', 'std', 'var'
```

# Count plot 


```python
sns.countplot(x='sex', data=tips) # The number of males and females in the tips dataset
```

# Box plot

```python
sns.set_context('paper', font_scale=1.4)
sns.boxplot(x='day', y= 'total_bill', data= tips, hue= 'sex') # Compar different variables and show quartiles 
plt.legend(loc= 0)
```

# Violin plot 

Similar to the box plot

```python
sns.violinplot(x='day', y='total_bill', data= tips, hue= 'sex', split= False)
```

# Strip plot

It plots a scater plot for different attributes grouped by the categorical attribute

```python
plt.figure(figsize=(8,5))
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex', dodge=True)
```

# Swarm plot

```python
sns.violinplot(x='day', y = 'total_bill', data = tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color = 'white')
```

# Palettes

```python
sns.set_context('talk')
sns.stripplot(x='day', y = 'total_bill', data= tips, hue= 'sex',)
plt.legend(loc=0)
```

```python
# Add a pattern 
plt.figure(figsize=(8,5))
sns.set_style('ticks')
sns.stripplot(x='day', y = 'total_bill', data=tips, hue = 'sex', palette='afmhot') 
# Palettes is basically a color map
plt.legend(loc= 'best')
```

# Pair grid plot

```python
iris = sns.load_dataset('iris')
plt.figure(figsize = (12,9))
sns.set_context('paper')
iris_g = sns.PairGrid(iris, hue= 'species')
#iris_g.map(plt.scatter)

# put a histogram plot along the diagonal
iris_g.map_diag(plt.hist)

# Remove the diagonal scatter plot 
iris_g.map_offdiag(plt.scatter)

# Offdiagonal plots 
iris_g.map_upper(sns.kdeplot)
iris_g.map_lower(plt.scatter)
iris_g.add_legend()
```

```python
plt.figure(figsize=(12, 12))
iris_g = sns.PairGrid(iris, hue='species', x_vars=['sepal_length', 'sepal_width'], y_vars=['petal_length', 'petal_width'])
iris_g.map(plt.scatter)
# Add legend 
iris_g.add_legend()
```

# Facete Grrid pot

```python
tipsfg = sns.FacetGrid(tips, col = 'time', row= 'smoker', height = 4, aspect=1.3)
tipsfg.map(plt.hist, 'total_bill', bins= 8) # Histogram plut 
```

```python
plt.figure(figsize=(12,8))
sns.set_context('talk')
tipsfg = sns.FacetGrid(tips, col='time', hue='smoker', height= 4, aspect=1.3, col_order = ['Dinner', 'Lunch'], palette='Set1')
tipsfg.map(plt.scatter, 'total_bill', 'tip')
```

```python
plt.figure(figsize=(12,8))
kws = {'s':70, 'linewidth':0.5, 'edgecolor': 'w'}
sns.set_context('paper')
tipsfg = sns.FacetGrid(tips, col='sex', hue='smoker', height= 4, aspect=1.3, hue_order = ['Yes', 'No'], hue_kws = dict(marker= ['o', 'v']))
tipsfg.map(plt.scatter, 'total_bill', 'tip', **kws)
tipsfg.add_legend()
```

```python
# load a students attention dataset 
plt.figure(figsize = (25, 19))
att = sns.load_dataset('attention')
attfg = sns.FacetGrid(att, col='subject', col_wrap=5, height=1.2)
attfg.map(plt.plot, 'solutions', 'score', marker = '*')
```

# Regrssion plot


```python
plt.figure(figsize=(8,6))
sns.set_context('paper', font_scale= 1.5)
kws = {'s':100, 'linewidth':0.5, 'edgecolor': 'w'}
sns.lmplot(x='total_bill', y = 'tip', hue = 'sex', data=tips, markers= ['o', '<'], scatter_kws=kws)
```
